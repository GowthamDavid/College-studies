```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def univariate_normal(x, mean, variance):
    return ((1 / np.sqrt(2 * np.pi * variance)) * np.exp(-(x - mean)**2 / (2* variance)))
```

```python
x = np.linspace(-3, 5, 100)
```

```python
fig = plt.figure(figsize = (5, 3))
```

```python
plt.plot(x,  univariate_normal(x, mean = 1, variance = 2), label = "$N(1,2)$")
plt.plot(x, univariate_normal(x, mean = 2, variance = 3), label = "$n(2,3)$")
plt.plot(x, univariate_normal(x, mean = 0, variance =  0.3), label = "$n(0, 0.3$")
plt.title("Univariate normal distribution")
plt.xlabel("X")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```

### Multivariate normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
mean = np.array([1, 2])
covariance_matrix = np.array([[1, 0.5], [0.5, 2]])
```

```python
x, y = np.meshgrid(np.linspace(-2, 4, 100), np.linspace(-2, 6, 100))
pos = np.dstack((x, y))
```

```python
pdf = multivariate_normal(mean, covariance_matrix)
```

```python
z = pdf.pdf(pos)
```

```python
plt.contourf(x, y, z, cmap='viridis')
plt.colorbar()  # Add a colorbar for reference
plt.title('Multivariate Gaussian Density Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
mean = np.array([2, 3])
covariance_matrix = np.array([[1, 0.7], [0.7, 4]])  # Non-diagonal covariance matrix
```

```python
# Create a grid of (x, y) values
x, y = np.meshgrid(np.linspace(-1, 5, 100), np.linspace(0, 6, 100))
pos = np.dstack((x, y))
```

```python
# Create a multivariate Gaussian PDF
pdf = multivariate_normal(mean, covariance_matrix)
```

```python
# Evaluate the PDF at each (x, y) point
z = pdf.pdf(pos)
```

```python
# Create a contour plot of the PDF
plt.contourf(x, y, z, cmap='viridis')
plt.colorbar()  # Add a colorbar for reference
plt.title('Correlated Multivariate Gaussian Density Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
